# ****************Covid 19 Rate in washington state. ***********************

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import dates as mpl_dates
import seaborn as sns
```

*****Read data from excel file *********

```python
data = pd.read_excel('PUBLIC_CDC_Event_Date_SARS.xlsx')
```

*******Finding out missing value **********

```python
missing_data = data.fillna(value='FILL VALUE')
```

```python
print(missing_data)
```

********** putting data in dataframe  *********

```python
df = pd.DataFrame(data)
df
```

************** maximum county corona virus in the NewPos_All. **************

```python
df.max()
```

************ Minimum County Corona virus in the NewPos_All

```python
df.min()
```

************ NewPos_All analysis **********

```python
print("Sum: ",df["NewPos_All"].sum()) 
print("Mean: ",df["NewPos_All"].mean())
print("Maximum: ",df["NewPos_All"].max())
print("Minimum: ",df["NewPos_All"].min()) 
```

***************** Top  maximum 5. *******************

```python
df.nlargest(5, 'NewPos_All')
```

************** Top Minimum 5.  ***********

```python
df.nsmallest(5, 'NewPos_All')
```

*************** Summarize NewPos_All corona virus by County  *************

```python
df_sub=df[["County","NewPos_All"]].groupby('County').sum()
df_sub
```

************ Analysing Age, NewPos_All and county **************

```python
df_sub=df[["County","Age 0-19","Age 20-39","Age 40-59","Age 60-79","Age 80+","NewPos_All"]].groupby('County').sum()
df_sub
```

********** Displaying rate of corona virus per county **********

```python
plt.figure(figsize=(20,5))
plt.plot(df['County'],df['NewPos_All'])
plt.plot(figsize=(20, 20))
plt.xlabel('NewPos_All')
plt.ylabel('County')
plt.title('NewPos_All vs County')


plt.show()
```
